## 1. DataLoading and Data PreProcessing

```python
import pandas as pd
from pandas import DataFrame
import numpy as np

movies=pd.read_csv("../python_workshop/data/movies.csv")
ratings=pd.read_csv("../python_workshop/data/ratings_1.csv")
movies.head()
```

```python
'''
ratings는 사용자별로 영화에 대한 평점을 매긴 데이터셋
평점은 최소 0.5~ 5점 사이에 0.5점 단위로 평점이 부여됨.
'''
```

```python
# 필요한 컬럼만으로 다시 DataFrame 만들기
ratings=ratings[["userId","movieId","rating"]]
'''
영화 title 필요 >> movie에서 ratings로 불러옴
'''
rating_movies=pd.merge(ratings,movies,on="movieId")
rating_movies
```

```python
# 1. 사용자를 row로 영화정보를  컬럼으로 구성한 피봇 생성
ratings_matrix = rating_movies.pivot_table('rating',index='userId',columns='title')
ratings_matrix.head(10)
```

```python
# 2. NaN값을 모두 0으로 변환
ratings_matrix=ratings_matrix.fillna(0)
ratings_matrix.head(10)
```

## 2. 영화 간의 유사도 산출 - Cosine Similarity
    사용자와 변환된 영화 평점 데이터 행렬 데이터셋을 이용해서 영화간 유사도를 측정해보자
    Cosine유사도는 기준이 되는 행과 다른 행을 비교해서 유사도를 측정하는 함수이다
    기존의 행렬 구조응 userId가 기준이 행 레벨의 데이터이기 때문에 사용자간의 유사도가 만들어진다.
    이를 사용자에서 영화간 유사도로 만들어보자.
    ※ Cosine유사도 사이킷런의 cosine_similarity() 함수를 사용

```python
# 아이템 기반의 유사도 측정을 위해 테이블을 전치
ratings_matrix_T =  ratings_matrix.transpose()
ratings_matrix_T.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

item_sim=cosine_similarity(ratings_matrix_T, ratings_matrix_T)
item_sim_df=pd.DataFrame(item_sim,index=ratings_matrix.columns, columns=ratings_matrix.columns)
item_sim_df
```

```python
# Godfather 와 유사도가 가장 높은 영화 6개..
item_sim_df['Godfather, The (1972)'].sort_values(ascending=False).head(6)
```
