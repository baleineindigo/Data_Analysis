## Shifting
    - 시계열에서 데이터를 Lagging할 때 쓰임

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df=pd.read_csv('../data/AAPL.csv')
# df의 인덱스를 datetime 형식으로 바꿔서 index로 지정
df.index=pd.to_datetime(df.index)
df
```

```python
# shift
temp=np.asarray(df["Close"])
df.shift(1).head(5) # 1step만큼 shift
```

```python
# 1step만큼 shift
df.tail() 
```

```python
# 3step만큼 shift하고 tail을 보면 끝에서 3번째까지 행은 없어지고 끝에서 4번째 행이 마지막에 있음
df.shift(3).tail() 
```

```python
# 거꾸로 shifting하면 1개행이 올라감 
df.shift(-1).head() 
```
