## 파도 모양 그래프

```python
t=np.arange(0,12,0.001)
'''
기본으로 존재하는 figure(=일종의 그래프가 들어가는 영역)가 아닌 
새로운 figure를 생성하기 위해서 pt.figure() 함수 사용
'''

plt.figure(figsize=(8,5))
plt.plot(t,np.sin(t),"m",lw=5,label="sin") # lw : line width.. 기본이 3
plt.plot(t,np.cos(t),"pink",lw=5,label="cos")


plt.legend()
plt.xlabel("Time")
plt.ylabel("Amplitute",rotation=0)
plt.title("Sin&Cos Graph")
plt.show()
```

# Pandas 데이터를 이용한 Visualization

```python
# 1. dictionary로 데이터 작성
np.random.seed(100)
data={"a":np.arange(50),
     'c':np.random.randint(0,50,50),
     "d":np.random.randn(50)}
data
```

```python
data['b']=data['a']+10*np.random.randn(50)
data['d']=np.abs(data['d'])*100
data
```

```python
import pandas as pd
from pandas import DataFrame
```

```python
data=DataFrame(data)
data
```

### 산점도(Scatter)

```python
'''
산점도 그래프는 scatter() 함수 이용해서 그린다
x,y축에 해당하는 데이터들의 상관관계를 표시할 때 데이터가 어떻게 흩어져 있는가를 알기 위해서 사용된다.

x,y 2개의 축을 기준으로 데이터가 얼마나 퍼져있는지(분포도)를 알 수 있기 때문에 산포도/산점도라고 함.
'''
# plt.scatter([1,2,3,4],[10,20,30,40],s=[10,20,30,40],color=["magenta","yellow","green","orange"])
# plt.show()

plt.scatter([1,2,3,4],[10,20,30,40],s=[30,40,50,60],c=range(4),cmap="jet")
plt.colorbar()
plt.show()
```

```python
# (20,20) 보다 큰 점은 다른 색으로 칠하기
# 방법 1) 
plt.scatter("a","b",data=data[data['a']<25],c="green")
plt.scatter("a","b",data=data[25<=data['a']],c="magenta")
```

```python
# 방법 2) 
cuthalf=(data["a"]>25.0)& (data["b"]>30.0)
data['color']=np.where(cuthalf==True,"black","magenta")

# 방법 3) 
import seaborn as sns
sns.regplot(x=data["a"],y=data["b"],scatter_kws={"facecolors":data["color"]})
plt.title("Scatter Plot")
```

## Subplot으로 plot(),scatter(),bar() 동시에 이용하기 

```python
names=["groupA","groupB","groupC"] # x축으로 사용
values=[1,10,100] # y축으로 사용
```

```python
plt.figure(1,figsize=(9,3)) # figsize=(9,3) 영역이 첫번째라는 의미로 figsize앞에 1을 붙임

plt.subplot(131) # 131 : 1행 3열의 1번째 그래프
plt.bar(names,values)
plt.grid()

plt.subplot(132) # 132 : 1행 3열의 2번째 그래프
plt.scatter(names,values)
plt.grid()

plt.subplot(133) # 133 : 1행 3열의 3번째 그래프
plt.plot(names,values)
plt.grid(color="lightgray")
```
